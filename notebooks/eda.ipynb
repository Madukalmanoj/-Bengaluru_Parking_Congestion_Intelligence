# Parking Violation EDA

Exploratory analysis of Bengaluru Traffic Police parking violations (Jan–May).
Used to justify Congestion Impact Score weights in `config.py`.

```python
import sys
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(ROOT))

import config

sns.set_theme(style="whitegrid")
```

```python
from src.ingest import ingest

df = ingest()
print(f"Records: {len(df):,}")
df.head()
```

```python
vt = df.explode("violation_types").groupby("violation_types").size().sort_values(ascending=False)
vt.head(15).plot(kind="barh", title="Top Violation Types")
plt.tight_layout()
plt.show()
```

```python
df.groupby("hour_of_day").size().plot(kind="bar", title="Violations by Hour (IST)")
plt.xlabel("Hour")
plt.tight_layout()
plt.show()
```

```python
from src.hotspots import detect_hotspots
from src.scoring import compute_scores

clusters = detect_hotspots(df)
scored = compute_scores(clusters)
scored[["rank", "congestion_impact_score", "violation_count", "dominant_junction"]].head(10)
```
